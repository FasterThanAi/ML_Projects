```python
def extract_episode(txt):
  check=False
  Data=""
  for i in txt:
    if i==')':
      check=False
      break

    if check==True:
      Data += i
    if i=='(':
      check=True
  return Data
```

```python
df['Episodes']=df['Title'].apply(extract_episode)
```

```python
# timeStamp Starts after ')'
def extract_time(txt):
  check=False
  data=""
  for i in txt:
    if len(data)>18:
      check=False
      break
    if check==True:
      data += i
    if i==')':
      check=True
  return data
```

```python
df['Total Time']=df['Title'].apply(extract_time)
```

```python
df.head()
```

```python
# Extract Month From Total Time
from dateutil.relativedelta import relativedelta
from datetime import datetime
def calculate_total_months(txt):
  try:
    start_str, end_str = txt.split(' - ')
    start_date = datetime.strptime(start_str.strip(), '%b %Y')
    end_date = datetime.strptime(end_str.strip(), '%b %Y')
    r = relativedelta(end_date, start_date)
    return r.years * 12 + r.months + 1
  except:
    return None

df['Months']=df['Total Time'].apply(calculate_total_months)
```
