```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique() # this col is aslo qualitative variable
```

```python
#distplot
sns.histplot(df['total_bill'],bins=20) # default value=20
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True) #kde= kernel density estimator
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
# joint plot
sns.jointplot(x='total_bill', y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip',data=df,kind='reg')
```

```python
# pair plot important
sns.pairplot(df,hue='size',palette='rainbow') # totalbills , size , tips ka comparison bcz they are numerical
```

```python
#rugplot
sns.rugplot(df['total_bill'])
```
