matplotlib


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.title('Basic Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#subplot
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

Object Oriented Way
matplotlib allows the aspect ratio , DPI and figured size to be specified when the figure object is created. You can use the figsize and DPI keyboard arguments
figsize is a tuple of the width and height of the figure in inches dpi is the dot per inch(pixel per inch)

```python
#Access the canvas of the plot
fig=plt.figure(figsize=(12,8),dpi=220) # 1200 x 800 px
axis1=fig.add_axes([0.1,0.1,1,1]) # last 2 ones 80%. width 80% height
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.4,0.3,0.3]) # first one left margin and 2nd one right margin
axis2.plot(x,y)
```
