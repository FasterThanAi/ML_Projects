```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.txt',sep=';', header=None,names=['text','emotion'])
```

create  6 dict. for each emotion

```python
unique_emotion=df['emotion'].unique()
```

```python
emotion_num={}
i=0
for emo in unique_emotion:
  emotion_num[emo]=i;
  i+=1;
df['emotion']=df['emotion'].map(emotion_num)
```

```python
df
```

lower case conversion

```python
df['text']=df['text'].apply(lambda x:x.lower())
```

```python
df
```

removing punctutation

```python
import string

def remove_punch(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df
```

```python
df['text']=df['text'].apply(remove_punch)
```

```python
df
```

remove numbers

```python
def remove_num(txt):
  new =""
  for i in txt:
    if not i.isdigit():
      new+=i
  return new
df['text']=df['text'].apply(remove_num)
```

```python
df
```

remove url and links . by removing https , /

remove emoji and special char

```python
def remove_emoji(txt):
  new=""
  for i in txt:
    if i.isascii():
      new+=i
  return new

df['text']=df['text'].apply(remove_emoji)
```

```python
import nltk
```

corpus: collection of text . like a dataset of text
eg : folder of 10K twitter msg
Token:  each individual word or symbol in the text.

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt') # it is lib that tokenize
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words('english'))
```

```python
len(stop_words)
```

each sentence me se stop word removal

```python
df.loc[1]['text']
```

```python
def remove(txt):
  #words=word_tokenize(txt)
  words=txt.split()
  cleaned=[]
  for i in words:
    if not i in stop_words:
      cleaned.append(i)
  return ' '.join(cleaned)
```

```python
df['text']=df['text'].apply(remove)
```

```python
df.loc[1]['text']
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
X_test
```

text vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer #BOW
from sklearn.feature_extraction.text import TfidfVectorizer #TF IDF

bow_vectorizer = CountVectorizer()
# tfidf_vectorizer = TfidfVectorizer()
```

```python
x_train_bow=bow_vectorizer.fit_transform(X_train)
# x_train_tfidf=tfidf_vectorizer.fit_transform(X_train)
```

```python
x_train_bow
```

```python
x_test_bow=bow_vectorizer.transform(X_test)
# x_test_tfidf=tfidf_vectorizer.transform(X_test)
```

naive bayes and logistic regresion

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
nb_model=MultinomialNB()
```

```python
nb_model.fit(x_train_bow,y_train)
```

```python
predict_nb=nb_model.predict(x_test_bow)
print(accuracy_score(y_test,predict_nb))
# print(classification_report(y_test,predict_nb)
```

```python
predict_nb
```

TF IDF

```python
tfidf_vectorizer = TfidfVectorizer()
x_train_tfidf=tfidf_vectorizer.fit_transform(X_train)
x_test_tfidf=tfidf_vectorizer.transform(X_test)

nb2_model=MultinomialNB()
nb2_model.fit(x_train_tfidf,y_train)
```

```python
y_pred=nb2_model.predict(x_test_tfidf)
print(accuracy_score(y_test,y_pred))
```

Logistic reg on TFIDF

```python
from sklearn.linear_model import LogisticRegression
```

```python
logis_model=LogisticRegression(max_iter=1000)
logis_model.fit(x_train_tfidf,y_train)
```

```python
log_pred=logis_model.predict(x_test_tfidf)
print(accuracy_score(y_test,log_pred))
```

bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer

documents=[
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]
vectorizer=CountVectorizer(ngram_range=(2,2)) #bigram

X=vectorizer.fit_transform(documents)
print("Vocabulary",vectorizer.get_feature_names_out())
print("\nBOW Matrix\n",X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents=[
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]
vectorizer=TfidfVectorizer()

X=vectorizer.fit_transform(documents)
print("Vocabulary",vectorizer.get_feature_names_out())
print("\nBOW Matrix\n",X.toarray())
```
